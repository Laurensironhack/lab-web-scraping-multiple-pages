Create a Python list with the top ten FBI's Most Wanted names:

```python
url2 = 'https://www.fbi.gov/wanted/topten'
```

```python
response2 = requests.get(url2)
response2.status_code
```

```python
soup2 = BeautifulSoup(response2.content, "html.parser")
```

```python
soup2
```

```python
soup2.select("h3")[0].get_text(strip=True)
```

```python
top10=[]
```

```python
#first I tried num_iter2=len etc etc. , but that gave me 12 results (not being names of fugitives). As I know I should only
get 10 results I just chose range(10)

for i in range(10):
    top10.append(soup2.select("h3")[i].get_text(strip=True))
```

A list with the different kind of datasets available in Datasets UK

```python
url3 = 'https://data.gov.uk/'
```

```python
response3 = requests.get(url3)
response3.status_code
```

```python
soup3 = BeautifulSoup(response3.content, "html.parser")
```

```python
soup3.select("#main-content > div:nth-child(3) > div > ul h3")[2].get_text()
```

```python
datasets=[]
```

```python
num_iter3 = len(soup3.select("#main-content > div:nth-child(3) > div > ul h3"))

for i in range(num_iter3):
    datasets.append(soup3.select("#main-content > div:nth-child(3) > div > ul h3")[i].get_text())
```
